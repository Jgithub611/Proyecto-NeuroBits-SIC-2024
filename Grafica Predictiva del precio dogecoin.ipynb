```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px

# Data Sampling

data = {
    'Year': [2024, 2025, 2026, 2027, 2028, 2029, 2030],
    'Min Price': [0.133, 0.1873, 0.267, 0.392, 0.5769, 0.8269, 1.23],
    'Avg Price': [0.1379, 0.1928, 0.2747, 0.403, 0.597, 0.8508, 1.27],
    'Max Price': [0.1455, 0.2306, 0.3279, 0.4617, 0.6868, 0.9855, 1.46]
}

df = pd.DataFrame(data)

# Line plot
plt.figure(figsize=(10, 6))
plt.plot(df['Year'], df['Min Price'], label='Min Price', marker='o')
plt.plot(df['Year'], df['Avg Price'], label='Avg Price', marker='o')
plt.plot(df['Year'], df['Max Price'], label='Max Price', marker='o')
plt.title('Dogecoin Price Predictions (2024-2030)')
plt.xlabel('Year')
plt.ylabel('Price (USD)')
plt.legend()
plt.grid(True)
plt.show()


correlation_data = pd.DataFrame({
    'Price': [1.27, 1.46],
    'Adoption Rate': [0.85, 0.90],
    'Miner Reserves': [300, 400]
}).corr()

plt.figure(figsize=(8, 6))
sns.heatmap(correlation_data, annot=True, cmap='coolwarm')
plt.title('Correlation Heatmap for Dogecoin Price Factors')
plt.show()
```
