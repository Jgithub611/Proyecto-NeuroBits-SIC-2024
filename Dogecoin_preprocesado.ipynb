```python
import pandas as pd
import matplotlib.pyplot as plt

dataset = pd.read_csv('DOGE-USD.csv')
dataset.dropna(inplace=True)
dataset['Date'] = pd.to_datetime(dataset['Date'])
años = dataset['Date'].dt.year.unique()

años_dataset = [2017,2018,2019,2020,2021,2022,2023,2024]

for año in años_dataset:
    df1 = dataset[dataset['Date'].dt.year == año]
    plt.figure(figsize=(20,8))
    df1.groupby('Date')['Volume'].mean().plot(linewidth = 1.5, marker ='o', color='#cb9800')
    plt.xlabel('Date')
    plt.ylabel('Volume')
    plt.title(f"Date vs Volume of {año}")
```
