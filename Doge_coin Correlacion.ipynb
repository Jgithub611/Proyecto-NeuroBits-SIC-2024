```
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from google.colab import files

sns.set_style('darkgrid')

uploaded = files.upload()
file_name = list(uploaded.keys())[0]
doge_data = pd.read_csv(file_name)

doge_data['Date'] = pd.to_datetime(doge_data['Date'])

doge_data_clean = doge_data.dropna()
correlation_matrix = doge_data_clean[['Open', 'High', 'Low', 'Close', 'Adj Close', 'Volume']].corr()


plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.2f', linewidths=0.5)
plt.title('Grafica de Correlaci√≥n para la moneda Dogecoin')
# Esto ayuda con la mejora de la visualizacion con etiquetas mas grandes (lo vi en java)
plt.xticks(fontsize=14, rotation=45, ha='right', color='black')
plt.yticks(fontsize=14, rotation=0)
plt.show()
```
